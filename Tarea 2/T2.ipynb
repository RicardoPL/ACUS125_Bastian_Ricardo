Usando la segunda ley Newton: 
 \begin{align*}
\sum_{}^{}F&=m\ddot{x}(t)
\end{align*}
\begin{align*}
m\ddot{x}(t)&=-c\dot{x}(t)-k\,x(t)
\end{align*}
\begin{align*}
m\ddot{x}(t)+c\dot{x}(t)+kx(t)&=0
\label{eq1}\tag{1}
\end{align*}

Donde la fuerza inercial es:
\begin{align*}
m\ddot{x}(t)
\end{align*}
El amortiguamiento en este caso de tipo viscoso:
\begin{align*}
c\dot{x}(t)
\end{align*}
y la fuerza elástica:
\begin{align*}
kx(t)
\end{align*}

Si dividimos por $m$ la Ec(1), queda:
\begin{align*}
\ddot{x}(t)+\frac{c}{m}\,\dot{x}(t)+\omega_{n}^{2}\,x(t)&=0
\label{eq2}\tag{2}
\end{align*}

Donde se define $\omega_{n}$:
\begin{align*}
\omega_{n}&=\sqrt{\frac{k}{m}},\quad {\text{frecuencia angular natural de movimiento}}\,\bigg( \frac{rad}{s}\bigg)
\end{align*}
y $\zeta$:
\begin{align*}
\zeta &=\frac{c}{2\,\sqrt{k\,m}},\quad {\text{razón de amortiguamiento (cantidad que no tiene unidades)}}
\end{align*}

Así, la ecuacion diferencial de movimiento en terminos de $\omega_{n}$ y $\zeta$:
\begin{align*}
\ddot{x}(t)+2\zeta \omega_{n}\,\dot{x}(t)+\omega_{n}^{2}\,x(t)&=0
\label{eq3}\tag{3}
\end{align*}

Usando las condiciones iniciales:
\begin{align*}
x(0)&=x_{0}
\end{align*}
\begin{align*}
\dot{x}(0)&=\dot{x}_{0}
\end{align*}

Una posible solución de Ec.(3):
\begin{align*}
x(t)&=Ae^{\alpha t}
\end{align*}
Reemplazando y derivando: 
\begin{align*}
\big(\alpha^{2}+2\zeta\omega_{0}\alpha+\omega_{0}^{2}\big)\,Ae^{\alpha\,t}&=0
\end{align*}

Por lo que:
\begin{align*}
\big(\alpha^{2}+2\zeta\omega_{0}\alpha+\omega_{0}^{2}\big)&=0
\end{align*} 
Resolviendo queda que $\alpha$: 
\begin{align*}
\alpha&= \omega_{0}\big(-\zeta\pm\sqrt{\zeta^{2}-1}\big)
\end{align*}

#### Asi entonces, estudiando todos los valores posibles que puede tomar $\alpha$:

### Caso 1: No amortiguado

Como $\zeta$ es 0, tenemos que:

\begin{align*}
\alpha =\pm j\omega_{n} 
\end{align*}

 La solucion general para este caso es:

\begin{align*}
x(t)=B_{1}e^{j\omega_{n}t}+B_{2}e^{-j\omega_{n}t} 
\end{align*}

> + $B_{1}$ y $B_{2}$ son constante de interpretación

 A partir de la solucion general y la formula de Euler, tenemos que:

\begin{align*}
x(t)=(B_{1}+B_{2})cos(\omega_{n}t)+j(B_{1}-B_{2})sen(\omega_{n}t)
\dot{x}(t)=-\omega_{n}C_{1}sen(\omega_{n}t)+\omega_{n}C_{2}cos(\omega_{n}t)
\end{align*}

> Si $C_{1}=B_{1}+B_{2}$ y $C_{2}=j(B_{1}-B_{2})$, entonces tenemos el vector posicion y vector velocidad:

\begin{align*}
x(t)=C_{1}cos(\omega_{n}t)+C_{2}sen(\omega_{n}t)
\label{eq1}\tag{a}
\end{align*}

\begin{align*}
\dot{x}(t)=-\omega_{n}C_{1}sen(\omega_{n}t)+\omega_{n}C_{2}cos(\omega_{n}t)
\label{eq2}\tag{a1}
\end{align*}

> Para hallar $C_{1}$ y $C_{2}$, tenemos que las condiciones iniciales son:

\begin{align*}
x(0)=x_{0};
\dot{x}(0)=\dot{x_{0}}
\end{align*}


>+ Entonces evaluando con las condiciones iniciales en la ecuación (a) y (a1), tenemos que:

\begin{align*}
x(0)=C_{1}=x_{0}
\end{align*}

\begin{align*}
\dot{x}(0)=\omega_{n}C_{2}=\dot{x}\Rightarrow C_{2}=\frac{\dot{x_{0}}}{\omega_{n}}
\end{align*}


Finalmente queda reemplazar, en la ecuacion (a), y tenemos la funcion que modela la vibración libre no amortiguada:

\begin{align*}
x(t)=x_{0}cos(\omega_{n}t)+\frac{\dot{x_{0}}}{\omega_{n}}sen(\omega_{n}t)
\end{align*}

```python
#EJEMPLO#
t=np.linspace(0,15,10000)

#DATOS
x_0=0 #Deplazamiento inicial (m)
V_0=30 #Velocidad inicial (m/s)
k=910 #(N/m)
m=50
Omega_n=np.sqrt(k/m)

##FUNCIÓN
x=x_0*np.cos(Omega_n*t)+(V_0/Omega_n)*np.sin(Omega_n*t)

#GRÁFICO
fig,ax=plt.subplots(nrows=1,ncols=1,figsize=(7,7),tight_layout=True)
ax.plot(t,x)
ax.set_xlabel('Tiempo',fontsize=16)
ax.set_ylabel('Amplitud',fontsize=16)
ax.set_title('No amortiguado',fontsize=20)
ax.grid(True,axis='y')
```

### Caso 2: $0<\zeta<1$

Si $0<\zeta<1$ , tenemos un caso de vibracion **sub-amortiguado**, y tiene dos raices, tales que:

\begin{align*}
\alpha =\omega_{n}\bigg[-\zeta\pm \sqrt{1-\zeta^{2}}\bigg]
\end{align*}
En este caso la solución general es: 
\begin{align*}
x(t)=B_{1}e^{(-\zeta\omega_{n}-j\omega_{n}\sqrt{1-\zeta^{2}})t}+B_{2}e^{(-\zeta\omega_{n}+j\omega_{n}\sqrt{1-\zeta^{2}})t}
\end{align*}
Desarrollando argebraicamente:
\begin{align*}
x(t)=e^{-\zeta\omega_{n}t}[B_{1}cos(\omega_{n}(\sqrt{1-\zeta^{2}})t)-jB_{1}sen(\omega_{n}(\sqrt{1-\zeta^{2}})t)+B_{2}cos(\omega_{n}(\sqrt{1-\zeta^{2}})t)+jB_{2}sen(\omega_{n}(\sqrt{1-\zeta^{2})t})]
\end{align*}
\begin{align*}
x(t)=e^{-\zeta\omega_{n}t}[(B_{1}+B_{2})cos(\omega_{n}(\sqrt{1-\zeta^{2}})t)-j(B_{2}-B_{1})sen(\omega_{n}(\sqrt{1-\zeta^{2}})t)]
\end{align*}
Siendo las nuevas constantes:
\begin{align*}
C_{1}=B_{1}+B_{2}
\end{align*}
\begin{align*}
C_{2}=j(B_{2}-B_{1})
\end{align*}
entonces:
\begin{align*}
x(t)=e^{-\zeta\omega_{n}t}[C_{1}cos(\omega_{n}(\sqrt{1-\zeta^{2}})t)+C_{2}sen(\omega_{n}(\sqrt{1-\zeta^{2}})t)]
\label{eq2}\tag{b}
\end{align*}

Usando las siguientes condiciones iniciales y calculando las constantes de integración:
\begin{align*}
x(0)&=x_{0}
\end{align*}
\begin{align*}
\dot{x}(0)&=\dot{x}_{0}
\end{align*}

Asi entonces,usando las siguientes condiciones iniciales y calculando las constantes de integración:
\begin{align*}
C_{1}=x_{0}
\end{align*}

\begin{align*}
C_{2}=\frac{\dot{x}_{0}+\zeta\omega_{n}x_{0}}{\omega_{n}\sqrt{1-\zeta^{2}}}
\end{align*}

Reemplazando en (b):
\begin{align*}
x(t)=e^{-\zeta\omega_{n}t}[x_{0}cos(\omega_{n}(\sqrt{1-\zeta^{2}})t)+\frac{\dot{x}_{0}+\zeta\omega_{n}x_{0}}{\omega_{n}\sqrt{1-\zeta^{2}}}sen(\omega_{n}(\sqrt{1-\zeta^{2}})t)]
\label{eqc}\tag{c}
\end{align*}

Como estrategia suponemos que la solución es:
\begin{align*}
x(t)&=Ae^{-\zeta\omega_{n}t}\,{\text{sen}}\big( \omega_{d}t+\phi_{d}\big)
\label{eqd}\tag{d}
\end{align*}
con:
\begin{align*}
\omega_{d}=\omega_{n}\sqrt{1-\zeta^{2}}
\end{align*}

Por lo que, igualando términos entre las ecuaciones (c) y (d),finalmente podemos definir:
\begin{align*}
A=\sqrt{x_0^{2}+(\frac{\dot{x}_{0}+\zeta\omega_{n}x_{0}}{\omega_{n}\sqrt{1-\zeta^{2}}})^{2}}
\end{align*}
\begin{align*}
\phi_{d}={\text{tan}}^{-1}\bigg( \frac{x_{0}\omega_{d}}{\dot{x}_{0}+\zeta\omega_{0}x_{0}} \bigg)
\end{align*}

```python
#EJEMPLO#
import numpy as np
from matplotlib import pyplot as plt
t=np.linspace(0,15,10000)
#Condiciones iniciales
x_0=0 #Deplazamiento inicial (m)
V_0=30 #Velocidad inicial (m/s)
c=100 #(kg/s)
k=910 #(N/m)
#--------------------------------------------------------------------------------------------------------------------------#
#Calculando términos para una masa 1
m1=50 #(Kg)
Omega_n1=np.sqrt(k/m1)
zeta1=c/(2*np.sqrt(k*m1))
A1=np.sqrt((x_0)**2+((V_0+(zeta1*Omega_n1*x_0))/(Omega_n1*(np.sqrt(1-(zeta1)**2))))**2)
Omega_d1=(Omega_n1)*np.sqrt(1-(zeta1)**2)
Phi_d1=np.arctan((x_0*Omega_d1)/(V_0+(zeta1*Omega_n1*x_0)))

#La función para una masa 1
x1=A1*np.exp(-zeta1*Omega_n1*t)*np.sin(Omega_d1*t+Phi_d1)

#--------------------------------------------------------------------------------------------------------------------------#
#Calculando términos para una masa 2
m2=100 #(Kg)
Omega_n2=np.sqrt(k/m2)
zeta2=c/(2*np.sqrt(k*m2))
A2=np.sqrt((x_0)**2+((V_0+(zeta2*Omega_n2*x_0))/(Omega_n2*(np.sqrt(1-(zeta2)**2))))**2)
Omega_d2=(Omega_n2)*np.sqrt(1-(zeta2)**2)
Phi_d2=np.arctan((x_0*Omega_d2)/(V_0+(zeta2*Omega_n2*x_0)))

#La función para una masa 2
x2=A2*np.exp(-zeta2*Omega_n2*t)*np.sin(Omega_d2*t+Phi_d2)

#--------------------------------------------------------------------------------------------------------------------------#
#Calculando términos para una masa 3
m3=150  #(Kg)
Omega_n3=np.sqrt(k/m3)
zeta3=c/(2*np.sqrt(k*m3))
A3=np.sqrt((x_0)**2+((V_0+(zeta3*Omega_n3*x_0))/(Omega_n3*(np.sqrt(1-(zeta3)**2))))**2)
Omega_d3=(Omega_n3)*np.sqrt(1-(zeta3)**2)
Phi_d3=np.arctan((x_0*Omega_d3)/(V_0+(zeta3*Omega_n3*x_0)))

#La función para una masa 3
x3=A3*np.exp(-zeta3*Omega_n3*t)*np.sin(Omega_d3*t+Phi_d3)
#--------------------------------------------------------------------------------------------------------------------------#
#GRÁFICO

fig,ax=plt.subplots(nrows=1,ncols=1,figsize=(7,7),tight_layout=True)
ax.plot(t,x1)
ax.plot(t,x2)
ax.plot(t,x3)
ax.set_xlabel('Tiempo',fontsize=16)
ax.set_ylabel('Amplitud',fontsize=16)
ax.set_title('Sub-amortiguado',fontsize=20)
ax.grid(True,axis='y')
```

### Caso 3: $\zeta=1$

Si $\zeta>1$ , tenemos un caso de vibración **críticamente-amortiguado**, y tiene una raiz, tales que:

\begin{align*}
\alpha =-\omega_{n}
\end{align*}
En este caso la solución de la ecuación diferencial es: $e^{-\omega_{n}t}$
La segunda solución linealmente independiente se obtiene al multiplicar la primera parte $t$, por lo que la solucion general es:
\begin{align*}
x(t)=e^{-\omega_{n}t}(C_{1}+C_{2}t)
\label{eqe}\tag{e}
\end{align*}

Usando las condiciones iniciales:

\begin{align*}
x(0)&=x_{0}
\end{align*}
\begin{align*}
\dot{x}(0)&=\dot{x}_{0}
\end{align*}

Se tiene que: 

\begin{align*}
C_{1}=x_{0}
\end{align*}
\begin{align*}
C_{2}=\dot{x}_{0}+\omega_{n}x_{0}
\end{align*}

Finalmente, reemplazando $C_{1}$ y $C_{2}$ en la ecuación (e)
\begin{align*}
x(t)=e^{-\omega_{n}t}(x_{0}+(\dot{x}_{0}+\omega_{n}x_{0})t)
\end{align*}

```python
###EJEMPLO
import numpy as np
from matplotlib import pyplot as plt
t=np.linspace(0,5,10000)
w_n=2
x_0=1
x1=-1
x2=10
x3=-15

#FUNCIONES
x_1=np.exp(-w_n*t)*(x_0+(x1+w_n*x_0)*t)
x_2=np.exp(-w_n*t)*(x_0+(x2+w_n*x_0)*t)
x_3=np.exp(-w_n*t)*(x_0+(x3+w_n*x_0)*t)

#GRÁFICO
from matplotlib import pyplot as plt 
%matplotlib inline
fig,ax=plt.subplots(nrows=1,ncols=1,figsize=(7,7),tight_layout=True)
ax.plot(t,x_1,t,x_2,t,x_3)
ax.set_xlabel('Tiempo',fontsize=16)
ax.set_ylabel('Amplitud',fontsize=16)
ax.set_title('Críticamente amortiguado',fontsize=20)
ax.grid(True,axis='y')
```

### Caso 4:  Si $\zeta>1$

Si $\zeta>1$ , tenemos un caso de vibración sobre-amortiguado, y tiene dos raices, tales que:

\begin{align*}
\alpha =\omega_{n}\bigg[-\zeta\pm \sqrt{1-\zeta^{2}}\bigg]
\end{align*}

La solución respecto a la ecuación caracteristica es:

\begin{align*}
x(t)&=e^{-\omega_{0}t}\bigg[C_{1}e^{-\omega_{n}\big(\sqrt{\zeta^{2}-1}\big)\,t}+C_{2}e^{\omega_{n}\big(\sqrt{\zeta^{2}-1}\big)\,t}\bigg]
\label{eq24}\tag{f}
\end{align*}

Aplicación de condiciones iniciales en la Ec. (f):
\begin{align*}
x(0)&=x_{0}
\end{align*} 
\begin{align*}
\dot{x}(0)&=\dot{x}_{0}
\end{align*}

>Entonces $C_{1}$ y $C_{2}$, queda definido como:

\begin{align*}
C_{1}=\frac{-x_{0}\omega_{n}(\zeta-\sqrt{\zeta-1}-\dot{x_{0})}}{2\omega_{n}\sqrt{\zeta-1}}
\end{align*}
\begin{align*}
C_{2}=\frac{\dot{x_{0}}+x_{0}\omega_{n}(\zeta+\sqrt{\zeta-1})}{2\omega_{n}\sqrt{\zeta-1}}
\end{align*}

Reemplando los terminos de $C_{1}$ y $C_{2}$ en la ecuación (f), se obtiene:

\begin{align*}
x(t)&=\frac{e^{-\omega_{0}t}}{2\big(\sqrt{\zeta^{2}-1}\big)}\,\cdot\bigg\{\,\bigg[\frac{\dot{x}_{0}}{\omega_{0}}+x_{0}\,\big(\zeta+\sqrt{\zeta^{2}-1}\,\big)\bigg]\,e^{\omega_{0}\big(\sqrt{\zeta^{2}-1}\big)\,t}\,+\,\bigg[-\frac{\dot{x}_{0}}{\omega_{0}}+x_{0}\,\big(-\zeta+\sqrt{\zeta^{2}-1}\,\big)\bigg]\,e^{-\omega_{0}\big(\sqrt{\zeta^{2}-1}\big)\,t}\,\bigg\}
\end{align*}

```python
#DATOS
w0= 3#(rad/s)
z= 1.2
x0=1 # 1 mm
v01=9
v02=20 # mm/s
t= np.linspace(0., 3.0, num=500)

#FUNCIONES
x1= (np.exp(-z*w0*t)/2*np.sqrt(z**2-1))*((v01/w0+x0*(z+np.sqrt(z**2-1)))*np.exp(w0*t*np.sqrt(z**2-1))+(-v01/w0+x0*(-z+np.sqrt(z**2-1)))*np.exp(-w0*t*np.sqrt(z**2-1)))
x2= (np.exp(-z*w0*t)/2*np.sqrt(z**2-1))*((v02/w0+x0*(z+np.sqrt(z**2-1)))*np.exp(w0*t*np.sqrt(z**2-1))+(-v02/w0+x0*(-z+np.sqrt(z**2-1)))*np.exp(-w0*t*np.sqrt(z**2-1)))

#GRÁFICO
fig,ax=plt.subplots(nrows=1,ncols=1,figsize=(7,7),tight_layout=True)
ax.plot(t,x1,t,x2)
ax.set_xlabel('Tiempo',fontsize=16)
ax.set_ylabel('Amplitud',fontsize=16)
ax.set_title('Sobre amortiguado',fontsize=20)
ax.grid(True,axis='y')
```
